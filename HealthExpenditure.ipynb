```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import os
```

### Extract CSVs into DataFrames

```python
expend_file_path = os.path.join("Data", "health_expend_cleaned.csv")
```

```python
expend_df = pd.read_csv(expend_file_path)
expend_df
```

```python
expend_df_clean = expend_df.melt(['Per.capita.government.expenditure.on.health.at.average.exchange.rate..US..'], var_name='Year', value_name='Value')
expend_df_clean
```

```python
expend_df_clean = expend_df_clean.replace(np.nan, '', regex=True)
```

```python
expend_df_clean
```

```python
expend_df_clean.to_csv("health_expend_cleaned.csv", encoding='utf-8', index=False)
```
