```python
import pandas as pd
import os
```

```python
#create filepath to CSV
file_path_obesity = os.path.join("Data", "obesity-cleaned - WHO.csv")
```

```python
#convert to pandas DataFrame
obesity_data = pd.read_csv(file_path_obesity)
obesity_data
```

```python
#use Melt to invert data 
obesity_data_melt =obesity_data.melt(['country_id'], var_name='year', value_name='value')
obesity_data_melt
```

```python
#set country_id as index
obesity_data_melt = obesity_data_melt.set_index(['country_id'])
obesity_data_melt
```

```python
#write to CSV
obesity_data_melt.to_csv("obesity_data_melt.csv")
```
